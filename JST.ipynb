```python
#! pip install keras
#Remove '#' in above line if not installed keras
import keras
import numpy as np
from sklearn.datasets import load_iris
dataset=load_iris()
#How data looks likes?
print (dataset)
```

```python
#Step : 1 : know the data
print(type(dataset))
print(len(dataset))
#Dataset is in bunch format. bunch is a dictionary like object.
print(dataset.keys())
#We try to find out what values are stored in these attributes
print(dataset['data'][0:5])
print(dataset['target'][0:5])
print(dataset['target_names'][0:5])
print(dataset['DESCR'])
print(dataset['feature_names'])
print(dataset['filename'])
#next we difine our x and y; y isthe dependent on x
x=dataset['data']
print(len(x)) #To check length of dataset
y=dataset['target']
print(len(y)) #To check length of dataset
print(x[0])
print(y[0])
```

```python
#Step : 2 : Convert y into one hot encoded vector
#One hot encoding is used because Y has labels (0,1,2)- which is a categorical categorical data whit no ordinal relationship
from keras.utils import to_categorical
Ny=len(np.unique(y))
print(Ny)
Y=to_categorical(y,num_classes=Ny)
print(Y[0:5])
```

```python
#Step3 : 3 : Now we split the data into two parts - one for training another for testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.10,shuffle=True)
print(x_train[0:5])
print(x_test[0:5])
print(y_train[0:5])
print(y_test[0:5])
```

```python
#Step : 4 : Then we normalize the data
from sklearn.preprocessing import StandardScaler
#Normalization is done so that the difference betwen highets and lowest data point is not too large
import numpy as np
scaler=StandardScaler()
#To find mean and std dev
scaler.fit(x_train)

#Conveting data into from where mean of data is 0 and std dev is 1
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
print(np.amax(x_train,axis=0))
print(np.amin(x_train,axis=0))
```

```python
#step : 5 : Now finally we bulid the model using keras
!pip install tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(20, input_dim=x_train.shape[1], activation='relu'))
#Dropout is used to avoid overfitting
keras.layers.Dropout(0.2)
model.add(Dense(20, activation='relu'))
keras.layers.Dropout(0.2)
model.add(Dense(20, activation='relu'))
keras.layers.Dropout(0.2)
#For classification problems, we usually use softmax as activition function function in final layey
model.add(Dense(3, activation='softmax'))
#Compiling the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Step : 6 : know model
model.summary()
```

```python
#Step : 7 : train the model
#Bath sixe,ec[ocs and validation split are all hyper parameters
model.fit(x_train, y_train, epochs=195, batch_size=80, validation_split=0.1)
```

```python
#Step : 8 : Get  the accuracy of model on testing data
testing=model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" %(model.metrics_names[1]+'uracy of model on testing data',testing[1]*100))
```

```python
#Step : 9 : Evaluate our model
from sklearn.metrics import classification_report, confusion_matrix
predictions = np.argmax(model.predict(x_test), axis=1)
y_test = np.argmax(y_test,axis=1)
#To get confusion matrix
print (confusion_matrix(y_test,predictions))
#To get values of all evaluation metrics)
print(classification_report(y_test,predictions))
```
